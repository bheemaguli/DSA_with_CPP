```cpp
#include<iostream>

using namespace std;

class Rectangle{
    private:
        int length, breadth;
    public:
        // Overloaded Constructor
        Rectangle(){length=breadth=0;}  // Default Constructor
        Rectangle(int l, int b);        // Paramererized Constructor

        // Facilitators
        int area();
        int perimeter();

        // Accessors OR Getter Function
        int getlength(){return length;}

        // Mutator OR Setter Function
        void setlength(int l){length = l;}

        // Destructorer
        ~Rectangle();
};

Rectangle::Rectangle(int l, int b){
    length = l;
    breadth = b;
}

int Rectangle::area(){
    return length*breadth;
}

int Rectangle::perimeter(){
    return 2*(length+breadth);
}

Rectangle::~Rectangle(){}

int main(){
    Rectangle r(18,3);
    cout<<"area : "<<r.area()<<endl;
    cout<<"perimeter : "<<r.perimeter()<<endl;
    r.setlength(20);
    cout<<"getlength : "<<r.getlength()<<endl;
}
```
