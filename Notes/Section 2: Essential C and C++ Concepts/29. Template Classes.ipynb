Consider,

```cpp

class Arthematic{
    private:
        int a;
        int b;
    public:
        Arthematic(int a, int b);
        int add();
        int sub();
};

Arithmetic::Arithmetic(int a, int b){
    this->a=a;
    this->b=b;
}

int Arithmetic::add(){
    int c;
    c = a + b;
    return c;
}

int Arithmetic::sub(){
    int c;
    c = a - b;
    return c;
}
```

Here, teh input supported isw only int type. If we want to use double as input and output, we'll have to create another class.

Instead of this, we'll create template which will specify structure of the class and this can be used for any suitable datatypes.

<br><br>

Templates are constucted as:

```cpp

template<class T>
class Arthematic{
    private:
        T a;
        T b;
    public:
        Arthematic(T a, T b);
        T add();
        T sub();
};
```

<br>

Template ends here. To define functions inside the template, we'll have to specify 'Template' again as:

```cpp

template<class T>
Arthematic<T>::Arthematic(T a, T b){
    this->a=a;
    this->b=b;
}

template<class T>
T Arthematic<T>::add(){
    T c;
    c = a + b;
    return c;
}

template<class T>
T Arthematic>T>::sub(){
    T c;
    c = a - b;
    return c;
}
```

<br><br>

Thus, the main function will be:

```cpp
int main(){
    Arthematic<int> ar{10,5};
    cout<<ar.add()<<endl;

    Arthematic<float> ar1{10.5,5.2};
    cout<<ar1.add()<<endl;
}
