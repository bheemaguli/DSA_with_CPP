# 49. Recurrence Relation - Time Complexity of Recursion

By Reconciliation:
Consider the following program:

```c
void fun1(int n){           - T(n)
    if(n>0){                - 1
        printf("%d",n);     - 1
        fun(n-1);           - T(n-1)
    }
}
```

The sum is 

$T(n) = T(n-1) + 2$

$T(n) â‰ˆ T(n-1) + 1$

The Total is:


$$
T(n)    = \left\{
    
        \begin{matrix}
        1                   & n=0 \\
        T(n-1) + 1   & n>0
        \end{matrix}
        \right.
$$

To solve the value of $T(n-1)$, lets use the induction method.

Here, if

$T(n) = T(n-1) + 1$

where $T(n-1) = T(n-1-1) + 1 = T(n-2) + 1$

Therefore,

$T(n) = T(n-2) + 1 + 1$

$T(n) = T(n-2) + 2$

Similarly,

$T(n) = T(n-3) + 1 + 2$

$T(n) = T(n-3) + 3$

<br>

For $k^{th}$ integer, 

$T(n) = T(n-k) + k$

<br>

Assume $n-k=0   =>  n=k$

Then,

$T(n) = T(n-n) + n$

$T(n) = T(0) + n$

$T(n) = 1 + n$

Which has the order $O(n)$

# 51. Static and Global Variables in Recursion

Static Variable declared inside a function is similar to global variable.

```c
int fun(int n){
    static int a=0;

    if(n>0){
        a++;
        return fun(n-1)+a;
    }
    return 0;
}

int main(){
    int x=5;
    printf("%d", fun(x));
    return 0;
}
```

will give same output as
