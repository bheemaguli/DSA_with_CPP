# Time Complexity

* $O(n)$

    If the operation is linear or one time execution without any inner loops like in array or single row of matrix.

    ```c
    for(int i=0;i<n;i++){
        ...
    }
    ```

    
* $O(n^2)$
    
    Case 1: If the operation has another operation duiring execution of one operation like two nested for loops where for each step in outer loop, there is $n$ execution by the inner for loop like in row and columns of matrix

    ```c
    for(int i=0;i<n;i++){
        for(int j=0;j<n;j++){
        ...
        }
    }
    ```

    Case 2: If first setp executes $n-1$ steps and the next step executes $n-2$, this will continue and will give
    
    $$(n-1) + (n-2) + (n-3)+ .... + 4 + 2 + 1 = \frac{n(n-1)}{2} = \frac{n^2 - n}{2}$$

    whose degree will be $n^2$

    ```c
    for(int i=0;i<n;i++){
        for(int j=i+1;j<n;j++){
        ...
        }
    }
    ```

* $O(log_{2}(n))$

    Case 1: If we are successly divide elements, we have time complexity of $log(n)$

    ```c
    for(int i=n;i>1;i=i/2){
        ...
    }
    ```
    OR

    ```c
    i=n;
    while(i>1){
        ...
        i = i/2;
    }
    ```

    Case 2: If we are operating/searching on a tree of height $n$, we will get time $O(log_{2}(n))$. If we are operating on all elements in a tree, then the complexity is $O(n)$

* $O(n^3)$

    If we have some operation taking $O(n)$ for each element in say a matrix, well have $n$ with degree of 3.

    ```c
    for(int i=0;i<n;i++){
        for(int j=0;j<n;j++){
            function(operation);
        ...
        }
    }
    ```

* $O(m + n)$

    If there is opreation taking $O(n)$ for $m$ elements, we will get above complixity. Here, depending on the magnitude of $m$ relative to $n$, the time complixity may change i.e., if $n>>m$ then time complixity is $O(n)$
