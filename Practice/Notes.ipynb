## 1.3 How Write and Analyze Algorithm
### 1.3.1 Rough sketch of algorithm

```c
    Algorithm swap(a, b){
        temp = a;
        a    = b;
        b    = temp;
    }
```

or

```text 

    Algorithm swap(a, b)
    Begin
        temp := a;
        a    := b;
        b    := temp;
    End
```

or 

```c

    Algorithm swap(a, b){
        temp <- a;
        a    <- b;
        b    <- temp;
    }
```

*There is no fixed method to write a rough working of algorithm.*



### 1.3.2 How to analyze an algorithm

**Criteria for Analysing Algorithm**
* Based on **Time**
* Based on **Space**
* Based on **Data Transfer / Network Consumption**
* Based on **Power Consumption**
* Based on **CPU Registers Consumption**

**How to Analyse Simple Algorithm**

Consider each statement in the below code takes 1 unit of time:
```c
    Algorithm swap(a, b){
        temp = a;           - 1
        a    = b;           - 1
        b    = temp;        - 1
    }                       -----
                              3
```                            

Here, running 3 statments will take 3 units of time when we do high level analysis.
Consider `x  = 5*a + 6*b`. Though we consider this as single statement, the code when converted to Low Level Language, multiple operations should occur just to execute the above one statement. For the scope of this course, we will not consider this Low Level Language Time Complexity.

> Personal Note: The above analysis reminds me of Andrej Karpathy tweet:[tweet:][id]
 [id]: https://twitter.com/karpathy/status/1430316576016793600 "Twitter"

<img src="ImSketches/tweet1.png" alt="drawing" width="200"/>

> Here, even for same operation of finding squate root, the backend implementation might vary from program to program and from library to library.
